```python
## Developer : Osanwuta Ogechukwu Jasper 
# Date : 18/10/2020
# Subject : Data Science: Python - DAT-5301 - MsBA1
# Program name : Escape Room


"""
    DocString:
    
    A) Introduction:
    Part of this game is based on the movie Escape Room (2019).
    It consists of three stages, and has defined function from
    escape room 1 to 3. Stage 1 to 2 requires you to think 
    differently and think fast and take note of your surrounding
    and stage 3 will suprise you and probably make you laugh.
    
    Round 1: Escape room 1
    Round 2: Escape room 2 
    Round 3: Escape room 3
    
    B) Known Bugs and/or Errors:
    None.
    
"""






# Global imports 
from sys import exit
import random
import time 

# Define functions
def room_1 ():
    global host_name
    global player_name 

def fail ():
    room_1()
    
    
def fail_2 ():
    print ("You're shot in the neck with a sleeper dart and...")
    room_1()

    
    
# Input function to put name of choice    
player_name = input (prompt = "Enter your name:\n")
host_name = """Mr. Miyagi"""

# Defining room 1
def room_1():
    # Printing story and dialogue between host and player 
    print ("""\n...You wake up in a well furnished bedroom, you look around...""")
    time.sleep(2)
    print ("\nYou're confused...")
    time.sleep(3)
    print ("\nYou quickly get up and try to break out by banging the door repeatedly")
    time.sleep(3)
    print ("\nThe Television comes on and you here a voice")
    time.sleep(2)
    print (f"""\n{player_name}: Where am I?! 
\n{host_name}: Welcome, you have been selected as a test subject in our new game...
ESCAPE ROOM!\n\nTo escape succefully, you'll have to play the game!
Try to escape the wrong way.....\nYou Die!""")
   
    print ("\nDo you wish to play the game?\nyes or no")
    
   # Player inputs choice "yes or no"
    choice_1 = input ("> ").lower()


    # Using nested conditional
    if "yes" in choice_1 or "y" in choice_1:
        time.sleep(2)
        escape_room_1() #This moves player to another room
   
    elif "no" in choice_1 or "n" in choice_1:
        time.sleep(2)
        fail()
   
    else:
        print ("\nWrong Answer!. Try Again")
        room_1() #Brings player back to room 1 to try again



# defining the escape_room_1   
def escape_room_1():
    time.sleep(2)
    print ("\nThere's a screw driver and a key underneath the pillow")
    time.sleep(2)
    
    print(f"""\n{player_name}: What do I do with this! Let me out!

{host_name}: Carefull, you know the rules. Ha Ha Ha Ha!. Use what you have to escape the room. Tick Tock {player_name}""")
    time.sleep(3)
    
    print ("\nThe room starts to fill up with water")
    time.sleep(6)
    
    print ("""\nHow do you wish to escape?
           1) use key to open the door and escape
           2) use the screw driver to unscrew the nuts holding the vent and \escape""")
    
    # Player inputs choice "1 or 2" OR any key word 
    choice_2 = input ("> ")
   
    
    # Using nested conditional 
    if "1" in choice_2 or "key" in choice_2:
        time.sleep(3)
        print ("\nMore water comes into the room and you drown\n")
       
        # Player Dies # 
        fail ()
    
    elif "2" in choice_2 or "screw" in choice_2 or "screw driver" in choice_2:
        time.sleep(3)
        escape_room_2() # This moves player to another room
        


# Defining escape room 2 
def escape_room_2():
    print ("\nYou escaped from the room 1 through the vents")
    time.sleep(3)
    print ("\nYou enter a room filled with books")
    time.sleep(3)
    print (f"\n{host_name}: Welcome to escape room 2")
    time.sleep(3)
    
    # Creating a while loop where the host asks questions
    print(f"""\n{host_name}: Answer my questions correctly and you progress,\answer incorrectly, and you start again
          PAY CLOSE ATTENTION in this stage, you'll need it.
          Ha Ha Ha!""")
    time.sleep(3)
    print ("\nQuestion 1")
    chances = 3
    
    # Using while loop and nested conditionals
    while chances > 0:
            choice_3 = input(prompt = """I'm as hard as a rock, but I melt immediately in hot water.
What am I?\nHere's a HINT, The rock is made of water\nI'm not supposed to be helping!\n""").lower()
            if "ice" in choice_3:
                print ("\nCorrect")
                question_2_loop() # this moves player to another while loop
                                  #where host asks another question
           
            elif choice_3 != "ice":
                chances -= 1
                if chances > 0:
                    print (f"\nThat's incorrect. You have {chances} chances remaining")
                    
                elif chances == 0:
                    print ("\nYou are out of guesses. The answer was Ice cube")
                    escape_room_2()
                    
            else:
                print ("\nSomething went wrong")


# Host is asking another question   
def question_2_loop():
    print ("\nQuestion 2")
    chances = 3 
    
    # Using while loop and nested conditionals
    while chances > 0:
        choice_4 = input (prompt = """What kind of room are you in?\n""").lower()
        if choice_4 == "library":
            print ("\nCorrect")
            escape_room_3() # This moves player to another room
            
        elif choice_4 != "library" or choice_4 != "Library":
            chances -= 1
            if chances > 0:
                print (f"\nThat's incorrect. You have {chances} chances remaining")
                
            elif chances == 0:
                print("\nYou are out of guesses. The answer was Library\n")
                fail_2()
                    
        else:
            print ("\nSomething went wrong") 
            
            
# Defining escape room 3 
def escape_room_3():
    print (f"""{host_name}: WELCOME! To the final escape room
           I must say, I am suprised you made it this far
           I'm sure you're itching to go outside""")
    time.sleep(4)
    print ("Final Challenge")
    time.sleep(3)
    print ("Tell us your favourite movie")
    time.sleep(3)
    print (f"{host_name}: Suprised eh! C'mon")
    time.sleep(3)
    # asking user to input favourite movie
    fav_mov = input(prompt = "What's your favourite movie?\n")
    time.sleep(2)
    # Use of dynamic string 
    print(f"{host_name}: {fav_mov} is my favorite movie too!")
    time.sleep(2)
    print (f"Congratulations {player_name}, you have successfully completed the escape room")
    play_again = input (prompt = "Do you wish to play again?\n yes or no\n").lower()
    
    # Nested conditional
    if "yes" in play_again:
        room_1()
    else:
        print ("Thanks for playing the escape room")
        exit (0)
        
        
# Room 1#        
room_1()            
            
```
